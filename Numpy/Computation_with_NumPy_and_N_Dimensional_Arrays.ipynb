#### N-Dimensional Arrays (Tensors)


```python
mystery_array = np.array([[[0, 1, 2, 3],
                           [4, 5, 6, 7]],
                        
                         [[7, 86, 6, 98],
                          [5, 1, 0, 4]],
                          
                          [[5, 36, 32, 48],
                           [97, 0, 27, 18]]])

# Note all the square brackets!
```

```python
print(f'We have {mystery_array.ndim} dimensions')
print(f'The shape is {mystery_array.shape}')
```

```python
# Access the value 18 
print(mystery_array[2, 1, 3])
```

```python
# Access the value [97, 0, 27, 18]
print(mystery_array[2, 1, :])
```

```python
# Retrieve a (3,2) matrix with the values [[ 0, 4], [ 7, 5], [ 5, 97]]
mystery_array[:, :, 0]
```

### Matrix Multiplication with @ and .matmul()

<img src=https://i.imgur.com/LCdhmi8.png width=350>

```python
a1 = np.array([[1, 3],
               [0, 1],
               [6, 2],
               [9, 7]])

b1 = np.array([[4, 1, 3],
               [5, 8, 5]])

print(f'{a1.shape}: a has {a1.shape[0]} rows and {a1.shape[1]} columns.')
print(f'{b1.shape}: b has {b1.shape[0]} rows and {b1.shape[1]} columns.')
print('Dimensions of result: (4x2)*(2x3)=(4x3)')
```

Let's multiply `a1` with `b1`.

```python
a1 @ b1
```

```python
print(np.matmul(a1, b1))
```

# Manipulating Images as ndarrays


```python
img = misc.face()
```

```python
plt.imshow(img)
```

What is the data type of `img`? Also, what is the shape of `img` and how many dimensions does it have? What is the resolution of the image?

```python
type(img)
```

```python
img.shape
```

```python
img.ndim
```

Convert the image to black and white. The values in our `img` range from 0 to 255. 

```python
sRGB_array = img / 255
```

```python
grey_vals = np.array([0.2126, 0.7152, 0.0722])
```

```python
img_gray = sRGB_array @ grey_vals
```

```python
plt.imshow(img_gray, cmap='gray')
```

Manipulate the images by doing some operations on the underlying ndarrays


```python
plt.imshow(np.flip(img_gray), cmap='gray')
```

```python
plt.imshow(np.rot90(img_gray), cmap='gray')
```
